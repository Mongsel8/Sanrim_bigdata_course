```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,1)
y = np.arange(0,10,1)
z = x**2

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection="3d")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel("z")
plt.plot(x,y,z,color = 'red')
plt.show()
```
